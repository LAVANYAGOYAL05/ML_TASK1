```python
# ML TASK 1: Linear and Logistic Regression from Scratch
# Implementation using NumPy and Pandas

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris, fetch_california_housing
import seaborn as sns

# =============================================================================
# PART 1: LINEAR REGRESSION FROM SCRATCH
# =============================================================================

class LinearRegressionScratch:
    """Linear Regression implemented from scratch using gradient descent"""

    def __init__(self, learning_rate=0.01, iterations=1000):
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.weights = None
        self.bias = None
        self.cost_history = []

    def compute_cost(self, X, y, weights, bias):
        """Calculate Mean Squared Error cost function"""
        m = len(y)
        predictions = X.dot(weights) + bias
        cost = (1/(2*m)) * np.sum((predictions - y)**2)
        return cost

    def gradient_descent(self, X, y):
        """Perform gradient descent to learn weights and bias"""
        m, n = X.shape
        self.weights = np.zeros(n)
        self.bias = 0

        for i in range(self.iterations):
            # Forward pass
            predictions = X.dot(self.weights) + self.bias

            # Calculate gradients
            dw = (1/m) * X.T.dot(predictions - y)
            db = (1/m) * np.sum(predictions - y)

            # Update parameters
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

            # Store cost for plotting
            cost = self.compute_cost(X, y, self.weights, self.bias)
            self.cost_history.append(cost)

            if i % 100 == 0:
                print(f"Iteration {i}: Cost = {cost:.4f}")

    def fit(self, X, y):
        """Train the model"""
        self.gradient_descent(X, y)

    def predict(self, X):
        """Make predictions"""
        return X.dot(self.weights) + self.bias

    def calculate_r2_score(self, y_true, y_pred):
        """Calculate R² score"""
        ss_total = np.sum((y_true - np.mean(y_true))**2)
        ss_residual = np.sum((y_true - y_pred)**2)
        r2 = 1 - (ss_residual / ss_total)
        return r2

    def calculate_mse(self, y_true, y_pred):
        """Calculate Mean Squared Error"""
        return np.mean((y_true - y_pred)**2)


# =============================================================================
# PART 2: LOGISTIC REGRESSION FROM SCRATCH
# =============================================================================

class LogisticRegressionScratch:
    """Logistic Regression implemented from scratch using gradient descent"""

    def __init__(self, learning_rate=0.01, iterations=1000):
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.weights = None
        self.bias = None
        self.cost_history = []

    def sigmoid(self, z):
        """Sigmoid activation function"""
        return 1 / (1 + np.exp(-z))

    def compute_cost(self, X, y, weights, bias):
        """Calculate binary cross-entropy cost function"""
        m = len(y)
        z = X.dot(weights) + bias
        predictions = self.sigmoid(z)

        # Avoid log(0)
        epsilon = 1e-5
        predictions = np.clip(predictions, epsilon, 1 - epsilon)

        cost = -(1/m) * np.sum(y * np.log(predictions) + (1-y) * np.log(1-predictions))
        return cost

    def gradient_descent(self, X, y):
        """Perform gradient descent to learn weights and bias"""
        m, n = X.shape
        self.weights = np.zeros(n)
        self.bias = 0

        for i in range(self.iterations):
            # Forward pass
            z = X.dot(self.weights) + self.bias
            predictions = self.sigmoid(z)

            # Calculate gradients
            dw = (1/m) * X.T.dot(predictions - y)
            db = (1/m) * np.sum(predictions - y)

            # Update parameters
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

            # Store cost for plotting
            cost = self.compute_cost(X, y, self.weights, self.bias)
            self.cost_history.append(cost)

            if i % 100 == 0:
                print(f"Iteration {i}: Cost = {cost:.4f}")

    def fit(self, X, y):
        """Train the model"""
        self.gradient_descent(X, y)

    def predict_proba(self, X):
        """Predict probabilities"""
        z = X.dot(self.weights) + self.bias
        return self.sigmoid(z)

    def predict(self, X, threshold=0.5):
        """Make binary predictions"""
        return (self.predict_proba(X) >= threshold).astype(int)

    def calculate_accuracy(self, y_true, y_pred):
        """Calculate accuracy"""
        return np.mean(y_true == y_pred)

    def confusion_matrix(self, y_true, y_pred):
        """Calculate confusion matrix"""
        tp = np.sum((y_true == 1) & (y_pred == 1))
        tn = np.sum((y_true == 0) & (y_pred == 0))
        fp = np.sum((y_true == 0) & (y_pred == 1))
        fn = np.sum((y_true == 1) & (y_pred == 0))

        return np.array([[tn, fp], [fn, tp]])


# =============================================================================
# LINEAR REGRESSION IMPLEMENTATION - California Housing Dataset
# =============================================================================

print("="*80)
print("LINEAR REGRESSION - California Housing Price Prediction")
print("="*80)

# Load dataset
housing = fetch_california_housing()
X_housing = pd.DataFrame(housing.data, columns=housing.feature_names)
y_housing = housing.target

print(f"\nDataset shape: {X_housing.shape}")
print(f"Features: {list(X_housing.columns)}")

# Split the data
X_train_lr, X_test_lr, y_train_lr, y_test_lr = train_test_split(
    X_housing, y_housing, test_size=0.2, random_state=42
)

# Standardize features
scaler_lr = StandardScaler()
X_train_lr_scaled = scaler_lr.fit_transform(X_train_lr)
X_test_lr_scaled = scaler_lr.transform(X_test_lr)

# Train Linear Regression
print("\nTraining Linear Regression Model...")
lr_model = LinearRegressionScratch(learning_rate=0.1, iterations=1000)
lr_model.fit(X_train_lr_scaled, y_train_lr)

# Make predictions
y_pred_train_lr = lr_model.predict(X_train_lr_scaled)
y_pred_test_lr = lr_model.predict(X_test_lr_scaled)

# Evaluate
train_mse = lr_model.calculate_mse(y_train_lr, y_pred_train_lr)
test_mse = lr_model.calculate_mse(y_test_lr, y_pred_test_lr)
train_r2 = lr_model.calculate_r2_score(y_train_lr, y_pred_train_lr)
test_r2 = lr_model.calculate_r2_score(y_test_lr, y_pred_test_lr)

print("\n" + "="*50)
print("LINEAR REGRESSION RESULTS")
print("="*50)
print(f"Training MSE: {train_mse:.4f}")
print(f"Testing MSE: {test_mse:.4f}")
print(f"Training R² Score: {train_r2:.4f}")
print(f"Testing R² Score: {test_r2:.4f}")

# Plot Loss vs Epoch for Linear Regression
plt.figure(figsize=(10, 5))
plt.plot(lr_model.cost_history)
plt.xlabel('Iteration (Epoch)')
plt.ylabel('Cost (MSE)')
plt.title('Linear Regression: Cost vs Iterations')
plt.grid(True)
plt.savefig('linear_regression_loss.png', dpi=300, bbox_inches='tight')
plt.show()

# Plot Predictions vs Actual
plt.figure(figsize=(10, 5))
plt.scatter(y_test_lr, y_pred_test_lr, alpha=0.5)
plt.plot([y_test_lr.min(), y_test_lr.max()], [y_test_lr.min(), y_test_lr.max()], 'r--', lw=2)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Linear Regression: Actual vs Predicted Prices')
plt.grid(True)
plt.savefig('linear_regression_predictions.png', dpi=300, bbox_inches='tight')
plt.show()


# =============================================================================
# LOGISTIC REGRESSION IMPLEMENTATION - Iris Binary Classification
# =============================================================================

print("\n" + "="*80)
print("LOGISTIC REGRESSION - Iris Binary Classification (Setosa vs Non-Setosa)")
print("="*80)

# Load Iris dataset
iris = load_iris()
X_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
# Binary classification: Setosa (0) vs Non-Setosa (1)
y_iris = (iris.target != 0).astype(int)

print(f"\nDataset shape: {X_iris.shape}")
print(f"Features: {list(X_iris.columns)}")
print(f"Class distribution: {np.bincount(y_iris)}")

# Split the data
X_train_log, X_test_log, y_train_log, y_test_log = train_test_split(
    X_iris, y_iris, test_size=0.2, random_state=42
)

# Standardize features
scaler_log = StandardScaler()
X_train_log_scaled = scaler_log.fit_transform(X_train_log)
X_test_log_scaled = scaler_log.transform(X_test_log)

# Train Logistic Regression
print("\nTraining Logistic Regression Model...")
log_model = LogisticRegressionScratch(learning_rate=0.1, iterations=1000)
log_model.fit(X_train_log_scaled, y_train_log)

# Make predictions
y_pred_train_log = log_model.predict(X_train_log_scaled)
y_pred_test_log = log_model.predict(X_test_log_scaled)

# Evaluate
train_acc = log_model.calculate_accuracy(y_train_log, y_pred_train_log)
test_acc = log_model.calculate_accuracy(y_test_log, y_pred_test_log)
cm = log_model.confusion_matrix(y_test_log, y_pred_test_log)

print("\n" + "="*50)
print("LOGISTIC REGRESSION RESULTS")
print("="*50)
print(f"Training Accuracy: {train_acc:.4f} ({train_acc*100:.2f}%)")
print(f"Testing Accuracy: {test_acc:.4f} ({test_acc*100:.2f}%)")
print(f"\nConfusion Matrix:")
print(cm)
print("\nConfusion Matrix Breakdown:")
print(f"True Negatives: {cm[0,0]}")
print(f"False Positives: {cm[0,1]}")
print(f"False Negatives: {cm[1,0]}")
print(f"True Positives: {cm[1,1]}")

# Plot Loss vs Epoch for Logistic Regression
plt.figure(figsize=(10, 5))
plt.plot(log_model.cost_history)
plt.xlabel('Iteration (Epoch)')
plt.ylabel('Cost (Binary Cross-Entropy)')
plt.title('Logistic Regression: Cost vs Iterations')
plt.grid(True)
plt.savefig('logistic_regression_loss.png', dpi=300, bbox_inches='tight')
plt.show()

# Plot Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Non-Setosa', 'Setosa'],
            yticklabels=['Non-Setosa', 'Setosa'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Logistic Regression: Confusion Matrix')
plt.savefig('logistic_regression_confusion_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

# Plot Decision Boundary (using first 2 features for visualization)
plt.figure(figsize=(10, 6))
X_vis = X_test_log_scaled[:, :2]  # Use first 2 features

# Create mesh
h = 0.02
x_min, x_max = X_vis[:, 0].min() - 1, X_vis[:, 0].max() + 1
y_min, y_max = X_vis[:, 1].min() - 1, X_vis[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

# Predict on mesh (using only first 2 features)
Z = log_model.sigmoid(
    np.c_[xx.ravel(), yy.ravel()].dot(log_model.weights[:2]) + log_model.bias
)
Z = Z.reshape(xx.shape)

# Plot
plt.contourf(xx, yy, Z, alpha=0.3, levels=np.linspace(0, 1, 11), cmap='RdYlBu')
plt.colorbar(label='Probability')
plt.scatter(X_vis[:, 0], X_vis[:, 1], c=y_test_log, cmap='RdYlBu',
            edgecolors='black', s=100, alpha=0.7)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title('Logistic Regression: Decision Boundary (First 2 Features)')
plt.savefig('logistic_regression_decision_boundary.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n" + "="*80)
print("IMPLEMENTATION COMPLETE!")
print("All graphs have been saved as PNG files.")
print("="*80)
```
